```python
# Calculate the correlation between runtime and rating
correlation = df['runtime'].corr(df['vote_average'])

# Display the correlation
print(f"Correlation between runtime and rating = {correlation:.2f}")

# Outcome: 
print('Indicates a very weak positive correlation between the two variables.')
```

## Part C

```python
# Create a histogram to visualise the distribution of movie ratings using plt.hist().

import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10,5) 
plt.hist(df['vote_average'], edgecolor='black')

plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()


# Outcome:
print('Insight: The distribution of movie ratings is approximately normal, with most ratings clustering between 5 and 7, peaking around 6, and fewer movies receiving extremely low or high ratings.')
```

```python
# Generate a scatter plot to show the relationship between runtime and rating using plt.scatter().

plt.scatter(df['runtime'], df['vote_average'], alpha=0.5)

plt.title('Relationship Between Runtime and Rating')
plt.xlabel('Runtime')
plt.ylabel('Rating')
plt.show()

# outcome:
print('Insight: The scatter plot shows that most movies have runtimes below 200 minutes with ratings clustered between 5 and 9, while longer movies are less frequent and exhibit a wider range of ratings.')
```

```python
# Create a bar chart to compare the average rating of movies in different genres using plt.bar().

genre_avg_ratings = df.groupby('genres')['vote_average'].mean().sort_values(ascending=False)

# Considering readability, keep only the top 5 genres by average rating
genre_avg_ratings = genre_avg_ratings.head(5)

# Plot bar chart
plt.figure(figsize=(20, 7)) # manipulating its size

plt.bar(genre_avg_ratings.index, genre_avg_ratings.values)
plt.xlabel('Genres')
plt.ylabel('Average Rating')
plt.title('Top Genres by Average Rating', fontsize=15)
plt.xticks(rotation=45, ha='right')
plt.show()

# Outcome: 
print('Insight: Drama|Horror|Mystery|Science Fiction|Thriller Category leads in the top movies in different genres by average rating')
```

## Data Visualization:

```python
# a. Use Matplotlib to create visualisations that highlight interesting patterns or insights from the data.

genre_avg_ratings = df.groupby('original_title')['revenue'].mean().sort_values(ascending=False)

# Considering readability, keep only the top 5 genres by average rating
genre_avg_ratings = genre_avg_ratings.head(5)

# Plot bar chart
plt.rcParams['figure.figsize'] = (20,7) 

plt.bar(genre_avg_ratings.index, genre_avg_ratings.values)

plt.xlabel('Movies')
plt.ylabel('revenue')
plt.title('Top 5 Movies by Revenue',fontsize=15)
plt.xticks(rotation=45, ha='right')
plt.show()

# Outcome:
print('Insight: Avatar is leading as the movie with that generates the highest revenue, followed by Star Wars:The Force Awakens, Jurassic World')
```

```python
# Comparing Average Popularity and Total Revenue by Movie Title 

df2 = df.head(10)
bar_width = 0.35

# Set up the figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))


# Positions for the bars
index = np.arange(len(df2['original_title']))

# Plot average popularity
bars1 = ax1.bar(index, df2['popularity'], bar_width, label='Average Popularity', color='blue')

# Create a second y-axis for total revenue
ax2 = ax1.twinx()
bars2 = ax2.bar(index + bar_width, df2['revenue'], bar_width, label='Total Revenue', color='brown')

# Add labels, title, and legend
ax1.set_xlabel('Movie Titles')
ax1.set_ylabel('Average Popularity', color='blue')
ax2.set_ylabel('Total Revenue', color='brown')
ax1.set_title('Comparison of Average Popularity and Total Revenue by Movie Title')
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(df2['original_title'], ha='right',rotation=50)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()

# Outcome:
print('Insight: Jurassic World and Star Wars: The Force Awakens stand out with both high popularity and total revenue, while Mad Max: Fury Road has comparatively high popularity but lower revenue, and the remaining films show varied levels of popularity and earnings.')
```

```python
# Comparing revenue across the released year 
plt.figure(figsize=(10, 6))  # Manipulating its size

# Create a line plot 
plt.plot(df['release_year'], df['revenue'], marker='o', linestyle='-', label='Yearly Rev')

plt.title('Yearly Revenue Trend', fontsize=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Rev', fontsize=12)
plt.legend()
plt.show()

print('Insight: There is an increase over time with some fluctuations, especially after the 1980s, where revenue growth becomes more pronounced with higher peaks, mostly around 2000s.')
```
